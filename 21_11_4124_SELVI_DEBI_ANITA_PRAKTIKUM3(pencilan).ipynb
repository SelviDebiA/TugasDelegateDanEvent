2. Buang data pencilan tersebut

```python
plt.figure(figsize=(5,5))
dataset.boxplot(['price'])
plt.show()
```

```python
Q1 = dataset['price'].quantile(0.50)
Q3 = dataset['price'].quantile(0.75)
```

```python
print(Q1, Q3)
```

```python
IQR = Q3 - Q1
```

```python
print(IQR)
```

```python
lower_limit = Q1 - 1.9 * IQR
upper_limit = Q3 + 1.9 * IQR

print(lower_limit,upper_limit)
```

```python
dataset[(dataset['price'] < lower_limit)]
```

```python
dataset[(dataset['price'] > upper_limit)]
```

```python
dataset[(dataset['price'] > lower_limit) & (dataset['price'] < upper_limit)]
```

```python
dataset_no_outlier = dataset[(dataset['price'] > lower_limit) & (dataset['price'] < upper_limit)]
```

```python
dataset_no_outlier
```
